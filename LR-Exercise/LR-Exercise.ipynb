```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("./canada_per_capita_income.csv")
df.head()
```

```python
df.tail()
```

```python
%matplotlib inline

plt.scatter(df['year'], df['per capita income (US$)'], c='r', marker="+")
plt.xlabel('Year (Date)')
plt.ylabel("Per capital Income")
plt.title("Date vs Per capital Income")
plt.show()
```

```python
model = linear_model.LinearRegression()
model.fit(df[['year']], df['per capita income (US$)'])
```

```python
y_predict = model.predict([[2020]])
y_predict
```

```python
plt.scatter(df['year'], df['per capita income (US$)'], c='r', marker="+")
plt.plot(df['year'], model.predict(df[['year']]), c="b")
plt.xlabel("Date")
plt.ylabel("per capital income")
plt.title("Actual vs Predicted")
plt.show()
```

```python
exercise_answer = model.predict([[2020]])
print(exercise_answer.round(2))
```
